# Experiment to use pandas as a 2d matrix to generate topic filters, like the visual keyword filter we used before.

```python
import pandas as pd
from sequemem import *
import numpy as np
```

```python
fpath = 'data/wonderwoman_topics.txt'
layer = Layer()
layer.train_from_file(fpath)
topics = []
with open(fpath, 'r') as source:
    for line in source:
        words = [word.strip() for word in line.split(' ')]
        if words[0] == 'about':
            topics.append(words[1])
topics        
```

```python
words = layer.column_keys()           
```

```python
idx = pd.Index(sorted(words), name='words')
df = pd.DataFrame(np.zeros((len(words),len(topics))), columns=sorted(topics), index=idx)
df.head()
```

```python
topic = None
with open(fpath, 'r') as source:
    for line in source:
        words = [word.strip() for word in line.split(' ')]
        if words[0] == 'about':
            topic = words[1]
            continue
        df.loc[list(set(words)), topic] += 1
        
df.head()
```
