```python

class Neuron:

    def __init__(self, parent_layer):
        self.ns_downstream = []
        self.state = 'I'
        self.ns_upstream = []
        self.keys = set()
        self.layer = parent_layer
        

    def add_key(self, key):
        self.keys.add(key)

    def get_keys(self):
        return self.keys

    def set_active(self):
        for neuron in self.ns_downstream:
            neuron.set_inactive()
        for neuron in self.ns_upstream:
            neuron.set_predict()

        if self.state == 'A':
            return
        elif self.state == 'I':
            self.layer.transition(self, "inactiv", "actives")
            self.state = 'A'
            return
        elif self.state == 'P':
            self.layer.transition(self, "predict", "actives")
            self.state == 'A'
            return
        assert False, "set_active failed"


    def set_predict(self):
        if self.state == 'P':
            return
        elif self.state == 'A':
            return
            #raise "active neuron can't be set to predict"
        elif self.state == 'I':
            self.layer.transition(self, "inactiv", "predict")
            self.state = 'P'
            return

        assert False, "set_predict failed"

    def set_inactive(self):
        if self.state == 'I':
            return
        elif self.state == 'P':
            self.layer.transition(self, "predict", "inactiv")
            self.state = 'I'
            return
        elif self.state == 'A':
            self.layer.transition(self, "actives", "inactiv")
            self.state = 'I'
            for neuron in self.ns_upstream:
                    neuron.set_inactive()
            return
        assert False, "set_inactive failed"

    def add_upstream(self, neuron):
        neuron.add_downstream(self)
        self.ns_upstream.append(neuron)

    def add_downstream(self, neuron):
        self.ns_downstream.append(neuron)

    def is_unused(self):
        return not self.ns_upstream and not self.ns_downstream
```

```python


class OutputLayer:
    def __init__(self):
        self.neurons = {}
        self.states = {
            "inactiv": set(),
            "actives": set(),
            "predict": set()
        }
        
    def transition(self, neuron, state_from, state_to):
        self.states[state_from].remove(neuron)
        self.states[state_to].add(neuron)
    
    def add_new(self, str_key):
        if str_key in self.neurons.keys():
            print("{} already exists".format(str_key))
            return
        else:
            neuron = Neuron(self)
            self.neurons[str_key] = neuron
            self.states["inactiv"].add(neuron)
            
    def get_neuron(self, str_key):
        return self.neurons[str_key]
    
    def set_active(self, lst_keys):
        for key in lst_keys:
            n = self.get_neuron(key)
            n.set_active()
    def active_keys(self):
        return list(set([k for neuron in self.states["actives"] for k in neuron.keys]))
            

alto = OutputLayer()
alto.add_new("A")
o = OutputLayer()
o.add_new("a")
o.add_new("a")
o.add_new("b")
type(o.get_neuron("a"))
n = o.get_neuron("a")
n.add_key("blb")
n.add_key("blb")
n.keys
n.state
print("\no neurons\n", o.neurons)
print("\no states\n", o.states)
print("\nalto neurons\n", alto.neurons)
print("\nalto states\n", alto.states)
n.set_active()
n.state
o.states["actives"]
o.active_keys()
n = alto.get_neuron("A")
nn = o.get_neuron("a")
nn.add_upstream(nn)
print("\no neurons\n", o.neurons)
print("\no states\n", o.states)
print("\nalto neurons\n", alto.neurons)
print("\nalto states\n", alto.states)

n.set_active()
print("\no neurons\n", o.neurons)
print("\no states\n", o.states)
print("\nalto neurons\n", alto.neurons)
print("\nalto states\n", alto.states)
```
